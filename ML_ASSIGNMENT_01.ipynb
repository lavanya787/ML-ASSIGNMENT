```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score
import graphviz
```

```python
X=df.drop('fetal_health',axis=1)
Y=df['fetal_health']
```

```python
plt.figure(figsize=(28,15))
plotnum=1

for col in X.columns:
  if plotnum<=9:
    ax=plt.subplot(5,5,plotnum)
    sns.boxplot(X[col])
    plt.xlabel(col)
  plotnum+=1
plt.show()
```

```python
#missing values
df.isna().sum
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.5,random_state=0)
```

```python
a=StandardScaler()
X_train = a.fit_transform(X_train)
X_test = a.transform(X_test)
```

```python
clf=DecisionTreeClassifier(criterion="entropy",random_state=0)
clf.fit(X_train,Y_train)
```

```python
Y_pred=clf.predict(X_test)
```

```python
cm=confusion_matrix(Y_test,Y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
from sklearn.tree import export_graphviz
```

```python
graphviz.Source(export_graphviz(clf,feature_names=X.columns,filled=True))
```

**BIAS AND VARIANCE PLOTTING**


```python
from sklearn.linear_model import LinearRegression
import datetime
from statistics import mean
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
pred = model.predict(X_test)
```

```python
#Variance
```

```python
Variance = np.var(pred)
Variance
```

```python
Sum_of_squared_error = np.mean((np.mean(pred)))
Bias = Sum_of_squared_error-Variance
Bias
```

```python
plt.scatter(Variance,Bias)
plt.show()
```
